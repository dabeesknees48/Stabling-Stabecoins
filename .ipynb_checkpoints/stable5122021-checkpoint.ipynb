```python
from cryptocmd import CmcScraper
import matplotlib.pyplot as plt
import numpy as np

#cmcscraper is free and so worth it saves you the headache of trying to create your own webparser since coinmarketcap doesnt do it for the free
scraper = CmcScraper("USDC")
usdc = scraper.get_dataframe()
scraper = CmcScraper("PAX")
pax = scraper.get_dataframe()
scraper = CmcScraper("BUSD")
busd = scraper.get_dataframe()
scraper = CmcScraper("GUSD")
gusd = scraper.get_dataframe()
```

```python
fig, (ax1, ax2) = plt.subplots(2)
fig.suptitle('GUSD')
ax1.plot(gusd['Date'], gusd['Volume'])
ax2.plot(gusd['Date'], gusd['Open']) #using open since cryptomarkets dont close it is merely just a marker
```

Gemini stablecoin looks rather stable and isn't . Still worth some stat tests.

```python
doffg = gusd['Open'] - 1
doffg.describe()
```

```python
doffg.skew()
```

```python
import statsmodels.tsa.stattools as sts
def stationary(ts,name):
#checks stationary with dicky fuller test#  
    fullerstats = sts.adfuller(ts)
    print('ADF Statistic: %f' % fullerstats[0])
    print('p-value: %f' % fullerstats[1])
    if (fullerstats[1] < .05):
        print (name + " Stationary")
    else:
        print (name + " Non-Stationary")

stationary(doffg,"Gemini")
```

```python
fig, (ax1, ax2) = plt.subplots(2)
fig.suptitle('PAX')
ax1.plot(pax['Date'], pax['Volume'])
ax2.plot(pax['Date'], pax['Open'])
```

Notice Pax isnt going thru histroical levels of trading

```python
doffp = pax['Open'] - 1
doffp.describe()
```

```python
doffp.skew()
```

```python
stationary(doffp,"Pax")
```
